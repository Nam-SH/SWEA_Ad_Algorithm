# 5186. 이진수2

<br>

![image](https://user-images.githubusercontent.com/50367487/64148516-72d01d00-ce5e-11e9-9ea0-25a6caf12507.png)

```python
def find(N):
    res = ''
    while N - int(N) != 0:
        N *= 2
        res += str(int(N))
        N -= int(N)
    if len(res) > 12: return 'overflow'
    return res

for tc in range(int(input())):
    N = float(input())
    print('#{} {}'.format(tc + 1, find(N)))
```
