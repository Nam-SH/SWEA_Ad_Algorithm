```python
for tc in range(1, int(input()) + 1):
    two = input().rstrip()
    three = input().rstrip()
    
    s2, s3 = set(), set()
    for i in range(1, len(two)):
        s2.add(int(two[:i] + str(1 - int(two[i])) + two[i + 1:], 2))
    
    s3.add(int('2' + three[1:] if three[0] == '1' else '1' + three[1:], 3))
    
    for i in range(1, len(three)):
        t = set('012')
        t.remove(three[i])
        for j in t:
            s3.add(int(three[:i] + j + three[i + 1:], 3))
    
    print("#{} {}".format(tc, s3.intersection(s2).pop()))
```

# 1240. 단순 2진 암호코드

<br>

![image](https://user-images.githubusercontent.com/50367487/64698499-fbddf880-d4dd-11e9-9171-4d845bbd1ee4.png)

```python
import sys; sys.stdin = open('data/(1240)input.txt', 'r')

dic = {'0001101': 0, '0011001': 1, '0010011': 2, '0111101': 3, '0100011': 4, 
       '0110001': 5, '0101111': 6, '0111011': 7, '0110111': 8, '0001011': 9}

for tc in range(int(input())):
    N, M = list(map(int, input().split()))
    arr = [input() for n in range(N)]

    for val in arr:
        if '1' in val:
            lst = val.rstrip('0')[-56:]
            break

    val_list = [dic[lst[i : i + 7]] for i in range(0, 50, 7)]

    odd = even = 0
    for i in range(0, len(val_list) - 1, 2):
        odd += val_list[i]
        even += val_list[i + 1]

    if (odd * 3 + even) % 10: res = 0
    else: res = odd + even
    print('#{} {}'.format(tc + 1, res))
```
